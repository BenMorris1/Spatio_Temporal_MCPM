```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, RationalQuadratic, Matern, ExpSineSquared, Sum, Product

np.random.seed(1)
```

## Functions

```python
# SPATIAL KERNELS
#################
# RBF Kernel
def rbf(x):
    return np.exp(-(x**2)/2)

# Rational Quadratic Kernel
def rq(x, alpha):
    return (1 + (x**2/(2*alpha)))**(-alpha)

# Matern-3/2 Kernel
def m32(x):
    return (1 + np.sqrt(3)*x)*np.exp(-np.sqrt(3)*x)

# Matern-5/2 Kernel
def m52(x):
    return (1 + np.sqrt(5)*x + (5*x**2)/3)*np.exp(-np.sqrt(5)*x)

# Exponential Kernel
def ex(x):
    return np.exp(-x)

# TEMPORAL KERNELS
##################
# Periodic Kernel                  
def periodic(x,p):
    return np.exp(-2*(np.sin(np.pi*x/p))**2)
                  
# Spectral Mixture Kernel
def sm(x):
    return np.exp(-0.5*x)*np.cos(x)            
```

## Create Input Data

```python
space_x = np.linspace(0.0,5.0,50)
time_x = np.linspace(0.0,2.0,50)
space_x,time_x = np.meshgrid(space_x,time_x)
inputs = np.array([space_x.flatten(),time_x.flatten()]).T

x1 = inputs[:,0]
x2 = inputs[:,1]
```

## RBF & Periodic

```python
# generate output data
z1 = rbf(x1)*periodic(x2,1)
z2 = rbf(x1)+periodic(x2,1)

# plot figure
fig  = plt.figure(figsize=(20,6))
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

ax1.plot_trisurf(x1, x2, z1, cmap=plt.cm.jet, linewidth=0)
ax1.set_xlabel("S - S'", fontsize='large')
ax1.set_ylabel("T - T'", fontsize='large')
ax1.set_zlabel('Value', fontsize='large')

ax2.plot_trisurf(x1, x2, z2, cmap=plt.cm.jet, linewidth=0)
ax2.set_xlabel("S - S'", fontsize='large')
ax2.set_ylabel("T - T'", fontsize='large')
ax2.set_zlabel('Value', fontsize='large')
```

## RBF & Spectral Mixture

```python
# generate output data
z1 = rbf(x1)*sm(x2)
z2 = rbf(x1)+sm(x2)

# plot figure
fig  = plt.figure(figsize=(20,6))
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

ax1.plot_trisurf(x1, x2, z1, cmap=plt.cm.jet, linewidth=0)
ax1.set_xlabel("S - S'", fontsize='large')
ax1.set_ylabel("T - T'", fontsize='large')
ax1.set_zlabel('Value', fontsize='large')

ax2.plot_trisurf(x1, x2, z2, cmap=plt.cm.jet, linewidth=0)
ax2.set_xlabel("S - S'", fontsize='large')
ax2.set_ylabel("T - T'", fontsize='large')
ax2.set_zlabel('Value', fontsize='large')
```
