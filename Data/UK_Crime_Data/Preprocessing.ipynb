```python
import pandas as pd
import numpy as np
import glob
import time
import datetime
```

```python
# Finding all 
path = 'Data/'
all_files = glob.glob(path + "/*.csv")
```

```python
# Getting crime counts for each space-time observation
data_final = pd.DataFrame()
for file in all_files:
    data = pd.read_csv(file)
    location = data['Reported by']
    date_time = data['Month']
    data_counts = pd.get_dummies(data['Crime type'])
    data_counts.iloc[0] = data_counts.sum(axis=0)
    data_counts = data_counts.head(1)
    data_counts['Location'] = location
    data_counts['Time'] = date_time
    data_final = pd.concat([data_final, data_counts], axis=0, ignore_index=True)
    
data_final
```

```python
# Getting unix time value for observations
def unix_time_calculator(x):
    year = int(x[:4])
    month = int(x[5:])
    day = 1
    d = datetime.datetime(year,month,day)
    unix_time = time.mktime(d.timetuple())
    return(unix_time)

data_final['Time'] = data_final['Time'].apply(lambda x: unix_time_calculator(x)) 
data_final
```
